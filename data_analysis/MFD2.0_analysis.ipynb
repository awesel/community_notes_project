```python
import pandas as pd
```

```python
file_path = "mfd2.0.dic"
hashmap = {}

with open(file_path, "r") as file:
    for line in file:
        line = line.strip()
        
        if not line or line.startswith("%"):
            continue
        
        phrase, header = line.split("\t")
        hashmap[phrase] = header

print(hashmap)
```

```python
words = []

with open(file_path, "r") as file:
    for line in file:
        line = line.strip()
        
        if not line or line.startswith("%"):
            continue
        
        phrase, header = line.split("\t")
        words.append(phrase)

print(words)
```

```python
# Read the combined CSV file
df = pd.read_csv("community_notes.tsv", sep='\t')

# Columns to be added with zero values
columns = ['care.virtue', 'care.vice', 'fairness.virtue', 'fairness.vice',
           'loyalty.virtue', 'loyalty.vice', 'authority.virtue', 'authority.vice',
           'sanctity.virtue', 'sanctity.vice']

# Add new columns with zero values
for column in columns:
    df[column] = 0

# Save the updated DataFrame to the final CSV file
df.to_csv("community_notes_m.csv", index=False)

print("done")
```

```python
df.dropna(subset=['summary'], inplace=True)
```

```python
def update_dataframe(row, words, hashmap):
    for word in words:
        if word in row['summary'] and not word.isdigit():
            word_key = hashmap[word]
            final_key = hashmap[word_key]
            row[final_key] += 1
            print(f"Just added one at {row['summary']} and {final_key}")
    return row

# Apply the custom function to each row in the DataFrame
df = df.apply(lambda row: update_dataframe(row, words, hashmap), axis=1)

# Save the modified DataFrame to a CSV file
df.to_csv("community_notes_m.csv", index=False)
```
