```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("community_notes.tsv", sep='\t')
```

```python
df.head()
```

```python
df = df[df['classification'] != 'NOT_MISLEADING']
```

```python
df.head(10)
```

```python
classification_unique = df['classification'].unique()
# Output the unique responses
print("---- classification_unique ----")
for response in classification_unique:
    print(response)

# -------

believable_unique = df['believable'].unique()
# Output the unique responses
print("---- believable_unique ----")
for response in believable_unique:
    print(response)

# -------

harmful_unique = df['harmful'].unique()
# Output the unique responses
print("---- harmful_unique ----")
for response in harmful_unique:
    print(response)
```

```python
df.shape
```

```python
df = df.dropna(subset=['harmful', 'believable'])
```

```python
df.shape
```

```python
# Calculate the frequency of each response in "harmful" and "believable" columns
harmful_counts = df['harmful'].value_counts()
believable_counts = df['believable'].value_counts()

# Create a pie chart for the "harmful" column
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.pie(harmful_counts, labels=harmful_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('harmful')

# Create a pie chart for the "believable" column
plt.subplot(1, 2, 2)
plt.pie(believable_counts, labels=believable_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('believable')

# Adjust layout for both charts
plt.tight_layout()

# Show the pie charts
plt.show()
```
